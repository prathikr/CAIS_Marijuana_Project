```python
# helper functions for displaying table data

import numpy as np
from IPython.display import display_html

# n is the number of columns to display data in
def display_side_by_side(series_obj, n):
    df = pd.DataFrame(series_obj)
    partition = int(round(len(df) / n))
    lower_bound = 0
    upper_bound = partition
    args = []
    for i in range(n):
        args.append(df[lower_bound:upper_bound])
        lower_bound += partition
        upper_bound += partition
    helper(args)

def helper(args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
import pandas as pd
pd.set_option('display.max_columns', 500)
import csv

# data: https://www.kaggle.com/muonneutrino/us-census-demographic-data#acs2017_county_data.csv
demographics = pd.read_csv("data/acs2017_county_data.csv")
demographics.head()
```

```python
import us

demographics['State_Code'] = demographics.apply(lambda x: us.states.lookup(x['State']).abbr, axis=1)
demographics.head()
```

```python
# data: https://www.kaggle.com/mikejohnsonjr/united-states-crime-rates-by-county
crime = pd.read_csv("data/crime_data_w_population_and_crime_rate.csv")
crime.head()
```

```python
cols_to_keep = ['county_name', 'crime_rate_per_100000', 'MURDER', 'RAPE', 'ROBBERY', 'BURGLRY', 'LARCENY',
               'ARSON']
crime = crime[cols_to_keep]
crime.head()
```

```python
# clean up crime county_name
temp = crime['county_name'].str.split(', ', expand=True)
crime['county_name'] = temp[0]
crime['state'] = temp[1]
crime.head()
```

```python
crime_and_demographics = crime.merge(demographics, left_on='county_name', right_on='County')
crime_and_demographics.head()
```

```python
# check states match to increase likelihood of correct mapping
crime_and_demographics = crime_and_demographics[crime_and_demographics.State_Code == crime_and_demographics.state]
crime_and_demographics.drop(columns=['State_Code'], inplace=True)
crime_and_demographics.head()
```

```python
# calculate percentage of NaNs in each column
percent_missing = crime_and_demographics.isnull().sum() * 100 / len(crime_and_demographics)
missing_value_df = pd.DataFrame({'column_name': crime_and_demographics.columns,'percent_missing': percent_missing})

display_side_by_side(missing_value_df, 2)
```

```python
crime_and_demographics.drop(columns=['ChildPoverty'], inplace=True)

# calculate percentage of NaNs in each column
percent_missing = crime_and_demographics.isnull().sum() * 100 / len(crime_and_demographics)
missing_value_df = pd.DataFrame({'column_name': crime_and_demographics.columns,'percent_missing': percent_missing})

display_side_by_side(missing_value_df, 2)
```
